<details>
<summary>Haz clic para ver la pista</summary>
Usa ".agg()" para aplicar funciones de promedio. Al final, reinicia el índice para mantener las columnas de la agrupación como variables (no índices).

````python

# ========================================================
# PASO 1: Crear traffic_city_year_2024 con .agg()
# ========================================================
# Diccionario de agregación con los NOMBRES EXACTOS de las columnas
diccionario_agg = {
    'jams_delay': 'mean',
    'traffic_index_live': 'mean',
    'jams_length_in_kms': 'mean',  # ← Con "in_kms" (no sin "in_")
    'jams_count': 'mean',
    'mins_delay': 'mean',
    'travel_time_live_per_10kms_mins': 'mean',
    'travel_time_historic_per_10kms_mins': 'mean'  # ← "historic" completo
}

traffic_city_year_2024 = (
    traffic_2024
    .groupby(['city', 'country', 'year'], as_index=False)
    .agg(diccionario_agg)
)

# ========================================================
# PASO 2: VERIFICACIONES Y VISUALIZACIÓN
# ========================================================
print("=== ¡Éxito! Dataset de tráfico promediado por ciudad ===")
print(f"Métodos de agregación aplicados:\n")
for col, metodo in diccionario_agg.items():
    print(f"  • {col}: {metodo}")

print(f"\n📊 Dataset final tiene {len(traffic_city_year_2024)} registros y {len(traffic_city_year_2024.columns)} columnas")
print(f"\nColumnas del resultado:\n{traffic_city_year_2024.columns.tolist()}\n")

print("=== TOP 10 CIUDADES CON MAYOR RETRASO (jams_delay) ===")
top_trafico = traffic_city_year_2024.sort_values("jams_delay", ascending=False).head(10)
display(top_trafico)

print("\n=== VISTA PREVIA DEL DATASET GENERAL (Primeras 5 filas) ===")
display(traffic_city_year_2024.head())

print(f"\n✅ Estructura del dataset:")
print(traffic_city_year_2024.info())

````

- Visualiza la distribución de la **economía** (`city_gdp_capita`) mediante:
    - **Histograma** → para analizar la forma de la distribución y el valor promedio del PIB per cápita.
- Finalmente, **compara ambas variables**, para observar si existe alguna relación entre ellas, haciendo un solo gráfico de barras donde aparezcan ambos indicadores.
- Recuerda agregar título y etiquetas a los ejes de tus gráficos.
- Observa y comenta los patrones, valores extremos o posibles relaciones que identifiques.

**Tip:** Dentro de los parentesis del boxplot, agrega `showmeans=True` para ver la media en el gráfico.

````python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Usar traffic_2024 y eco_2024 que tienen datos
merged = traffic_2024.merge(eco_2024, on=['city', 'year'], how='inner')

print(f"✅ Merged: {merged.shape[0]} registros\n")

fig, axes = plt.subplots(1, 3, figsize=(16, 4))

# BOXPLOT
axes[0].boxplot([merged['jams_delay']], showmeans=True)
axes[0].set_title('Boxplot - Jams Delay')
axes[0].set_ylabel('Minutos')

# HISTOGRAMA
axes[1].hist(merged['city_gdp_capita'], bins=15, color='steelblue', edgecolor='black')
axes[1].axvline(merged['city_gdp_capita'].mean(), color='red', linestyle='--', linewidth=2)
axes[1].set_title('Histograma - GDP per Cápita')
axes[1].set_xlabel('USD')

# BARRAS
sample = merged.sample(min(12, len(merged)))
x = np.arange(len(sample))
width = 0.35

traffic_norm = (sample['jams_delay'] - sample['jams_delay'].min()) / (sample['jams_delay'].max() - sample['jams_delay'].min()) * 100
gdp_norm = (sample['city_gdp_capita'] - sample['city_gdp_capita'].min()) / (sample['city_gdp_capita'].max() - sample['city_gdp_capita'].min()) * 100

axes[2].bar(x - width/2, traffic_norm, width, label='Tráfico', color='coral')
axes[2].bar(x + width/2, gdp_norm, width, label='GDP', color='steelblue')
axes[2].set_title('Comparación (Normalizado)')
axes[2].set_xticks(x)
axes[2].set_xticklabels(sample['city'].values, rotation=90, fontsize=8)
axes[2].legend()

plt.tight_layout()
plt.show()

# Análisis
print(f"Correlación: {merged['city_gdp_capita'].corr(merged['jams_delay']):.3f}")
print(f"\nTráfico - Media: {merged['jams_delay'].mean():.2f}, Mediana: {merged['jams_delay'].median():.2f}")
print(f"GDP - Media: ${merged['city_gdp_capita'].mean():.2f}")
````

**Tip:** Antes del `plt.show()` agrega el código `plt.xticks(rotation=90)` para rotar las etiquetas del eje X en 90 grados.
